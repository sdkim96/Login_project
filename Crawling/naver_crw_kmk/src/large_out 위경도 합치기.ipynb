```python
import pandas as pd
import numpy as np
```

```python
large1 = pd.read_csv('../DAOU/test_large_out_v1.csv', encoding='euc-kr')
large2 = pd.read_csv('../DAIN/crw_large_list_v2.txt', index_col=0)
large3 = pd.read_csv('../DAOU/test_large_out_v1.csv', encoding='euc-kr')
```

```python
large1.drop(large1.columns[0], axis=1, inplace = True)
large3.drop(large3.columns[0], axis=1, inplace = True)
large1.head()
```

```python
large3.info()
```

```python
#print(large2.head())
large2.info()
```

```python
print(large1[~np.isnan(large1['위도'])].shape)
print(large2[~np.isnan(large2['위도'])].shape)

#print(large1.head())
#print(large2.head())
```

```python
l1 = ~np.isnan(large1['위도'])
l2 = ~np.isnan(large2['위도'])
print(type(l1))
print(l1)
```

```python
l1_F_index = set(l1.index[l1 == False].tolist())
l2_T_index = set(l2.index[l2 == True].tolist())
print(l1_F_index)
print(l2_T_index)

l3_index = list(l1_F_index & l2_T_index)
print(l3_index)
print(len(l3_index))

large3.loc[l3_index,'위도'] = large2.loc[l3_index,'위도']
large3.loc[l3_index,'경도'] = large2.loc[l3_index,'경도']
```

```python
large3.to_csv("../DAOU/test_large_out_v2.csv", encoding='euc-kr')
large3.info()
```
